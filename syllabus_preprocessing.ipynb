```python
import pandas as pd
import camelot
```

```python
path = "input/syllabus/IBCompSciGuide.pdf"
tables = camelot.read_pdf(path, pages="all")
tables[0].df
```

```python
df = pd.DataFrame()
for i, table in enumerate(tables):
    print(f"Table {i+1}")
    df = pd.concat([df, table.df], ignore_index=True)
df.to_excel("input/syllabus/extracted/extracted.xlsx", header=False, index=False)
```

`extracted.xlsx` was cleaned manually to remove extra rows. The new file is called `filtered_syllabus.xlsx`. 

```python
syllabus_pd = pd.read_excel("input/syllabus/extracted/filtered.xlsx")
syllabus_pd
```

```python
syllabus_pd["Assessment Statement"] = (
    syllabus_pd["Assessment Statement"]
    .str.replace("\n", " ")
    .replace(r"\s+", " ", regex=True)
)
syllabus_pd["Teacher's Notes"] = (
    syllabus_pd["Teacher's Notes"]
    .str.replace("\n", " ")
    .replace(r"\s+", " ", regex=True)
)
syllabus_pd.set_index("Topic Number", inplace=True)
```
